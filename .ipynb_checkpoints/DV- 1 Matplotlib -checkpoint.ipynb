## Kütüphaneleri import edin

```python
import numpy as np
import pandas as pd

import matplotlib as mpl 
import matplotlib.pyplot as plt

%matplotlib inline
#%matplotlib notebook

import warnings;
warnings.filterwarnings("ignore")
```

## Değişkenlerimizi tanımlayalım

m değerini output a göre numpy kütüphanesi yardımı ile oluşturalım

```python
m = np.arange(0,11 , dtype = float)
m
```

```python
m
```

```python
c = 3* 10**8
```

```python
E = m* c**2
```

```python
E
```

## Oluşturduğumuz dokümanları plotlayalım

#### PLOTTING E=mc^2

```python
plt.figure(figsize = (4.3, 2.8))

plt.plot(m, E ,"b", lw = 3)
plt.title("E = mc^2", fontsize = 9)
plt.xlabel("M in g", fontsize = 9)
plt.ylabel("E in j", fontsize = 9)
```

```python
plt.figure(figsize = (5.6, 4.2))

plt.plot(m, E , "orange", lw = 2)
plt.title("E = mc^2", fontsize = 9)
plt.xlabel("M in g", fontsize = 9)
plt.ylabel("E in j", fontsize = 9)
```

## Alttaki görseli oluştururken farklı bir değişken tanımlamayın. Bence eksenlerden birinin logaritmik hesaplanması gerekebilir:)

## Ayrıca axes de oluşan gridler bir style değil bence matplotlib library de gridleri araştırmanız gerekir

```python
plt.figure(figsize = (4.5, 3))

plt.plot(m, E, "g",lw = 2)
plt.yscale("log")  


plt.title("E = mc^2", fontsize = 9)
plt.xlabel("M in g", fontsize = 9)
plt.ylabel("E in j", fontsize = 9)

plt.grid(True, which = "both", axis="y", linestyle = "-", linewidth = 0.8, alpha = 1)
```

## Aşağıdaki değişkenlere göre ilgili platlamaları yapın

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]

positions = range(len(labels))

plt.figure(figsize =(5, 4))
plt.plot(positions, july16_2007, label ="July 16, 2007")
plt.plot(positions, july16_2020,  label ="July 16, 2020")

plt.xticks(positions, labels, rotation = 0, fontsize = 8)
plt.yticks(fontsize = 8)

plt.legend(loc = "right")

plt.tight_layout()  
plt.show()
```

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]

positions = range(len(labels))

plt.figure()
plt.plot(positions, july16_2007, label = "July 16, 2007")
plt.plot(positions, july16_2020,  label = "July 16, 2020")

plt.xticks(positions, labels, rotation = 0, fontsize = 8)
plt.yticks(fontsize = 8)

plt.legend(bbox_to_anchor = (1.4, 0.5))

plt.tight_layout()  
plt.show()
```

## Alttaki görselleri araştırarak yapmaya çalışalım. Zihnimizi ve parmaklarımızı zorlayalım:)

#### Bir hatırlatma "fig, ax = plt.subplots" syntax i size cevabı bulmanızda yardımcı olur. 

#### ayrıca shift + tab+ tab yapmayı unutmayın ve içindeki parametreleri denemeyi unutmayın.

```python
fig, (ax1, ax2) = plt.subplots(2, 1)  

positions = range(len(labels))

ax1.plot(positions, july16_2007, "-", label = "July 16, 2007")
ax1.set_title("July 16, 2007 Yield Rates")
ax1.set_xticks(positions)
ax1.set_xticklabels(labels, rotation=0)

ax2.plot(positions, july16_2020, "-", label =  "July 16, 2020")
ax2.set_title("July 16, 2020 Yield Rates")
ax2.set_xticks(positions)
ax2.set_xticklabels(labels, rotation = 0)

plt.tight_layout()
plt.show()
```

## Bu görselleri alabilenler için alkışı borç bilirim

```python
fig, ax1 = plt.subplots(figsize = (10, 5))

ax1.plot(positions, july16_2007, "b-", label = "July 16, 2007")
ax1.set_ylabel("July 16, 2007", color = "b",fontsize = "15")

ax2 = ax1.twinx()

ax2.plot(positions, july16_2020,  "r-" ,label = "july16_2020")
ax2.set_ylabel("July 16, 2020", color = "r", fontsize = "15")


ax1.set_xticks(positions)

ax1.set_xticklabels(labels, rotation = 0)
ax1.legend ()
ax2.legend(loc = "right") # 2 axes ve  2 tane legend var diye extra yazmak istedim:)

fig.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.plot(positions, july16_2007, "b-", label = "July 16, 2007")
ax1.set_ylabel("July 16, 2007", color = 'b',fontsize = "15")
ax1.spines["left"].set_color("blue")
ax1.spines["left"].set_linewidth(1.5)

ax2 = ax1.twinx()

ax2.plot(positions, july16_2020,  "r-" ,label = "july16_2020")
ax2.set_ylabel("July 16, 2020", color = "r", fontsize = "15")


ax1.set_xticks(positions)
ax1.set_xticklabels(labels, rotation = 0)


ax2.spines["right"].set_color("red")

fig.tight_layout()
plt.show()
```
